```python
import pandas as pd
import matplotlib.pyplot as plt
htxdata = pd.read_csv('./data/houston.csv') # columns are year, city, country, avg_temp
globdata = pd.read_csv('./data/global.csv') # columns year, avg_temp
sanfrandata = pd.read_csv('./data/sanfran.csv')
```

```python
plt.plot(globdata['year'],globdata['avg_temp'])
plt.show()
# sweet nelly that's erratic
```

```python
plt.plot(htxdata['year'],htxdata['avg_temp'])
plt.show()
# sweet nelly thats erratic....er?
```

```python
htxMovingAverage= htxdata['avg_temp'].rolling(window=7,min_periods=7).mean()
globMovingAverage= globdata['avg_temp'].rolling(window=7,min_periods=7).mean()
sanfranMovingAverage= sanfrandata['avg_temp'].rolling(window=7,min_periods=7).mean()
# creating moving averages
# weekly averaging gives us a good idea of the overall climate for the region
```

```python
plt.plot(globdata['year'],globMovingAverage, label='Global Average')
plt.plot(htxdata['year'],htxMovingAverage, label='Houston, TX')
plt.plot(sanfrandata['year'],sanfranMovingAverage, label='San Francisco, CA')
plt.legend()
plt.ylabel('Temperature (C)')
plt.xlabel('Year')
plt.title('Annual Avg Temp vs Time')
plt.show()
# print(pd.DataFrame({'htxMA':htxMovingAverage, 'globMA':globMovingAverage}))
```

### Conclusions
    Houston's annual temperature is more than 10C higher each year than global averages over the entirety of the measurement record. This could likely be explained due to its proximity to the equator as well as the Gulf of Mexico.
    The overal trend for both global and Houston temperatures is towards a slightly warmer climate. One should not extrapolate using linear regression because we do not have data outside of the range to conjecture future values. We can, however, generalize that warming is likely to continue. It should be noted that while the timescale is long in terms of human lifespan, it is very miniscule in the context of geological phenomena. 
    As a result of the data being given to us as an average temperature, and then making a moving average of that temp, we are functionally taking averages of averages. This is statistically an issue in that we see both propogation of error in both calculations as well as biasing of the data. However, much like the linear regression, this is allowed for collecting a general idea of the trend of the data as opposed to making numerical calculations.
    The global averages seem to be unusually low, which is likely because it is averaging all data across the world for the entire year. In order to get a better idea I have also included the data for San Francisco, California. It is also on the North American continent and thus can at least give some context to the extremity of the Houston data.
