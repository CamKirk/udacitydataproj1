```python
import pandas as pd
import matplotlib.pyplot as plt
htxdata = pd.read_csv('./data/houston.csv') # columns are year, city, country, avg_temp
globdata = pd.read_csv('./data/global.csv') # columns year, avg_temp
```

```python
plt.plot(globdata['year'],globdata['avg_temp'])
plt.show()
# sweet nelly that's erratic
```

```python
plt.plot(htxdata['year'],htxdata['avg_temp'])
plt.show()
# sweet nelly thats erratic....er?
```

```python
htxMovingAverage= htxdata['avg_temp'].rolling(window=7,min_periods=7).mean()
globMovingAverage= globdata['avg_temp'].rolling(window=7,min_periods=7).mean()
# creating moving averages
```

```python
plt.plot(globdata['year'],globMovingAverage)
plt.plot(htxdata['year'],htxMovingAverage)
plt.show()
print(pd.DataFrame({'htxMA':htxMovingAverage, 'globMA':globMovingAverage}))
```
